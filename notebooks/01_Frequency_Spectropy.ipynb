```python
import numpy as np
import matplotlib.pyplot as plt
import qiskit, qiskit_experiments
from qiskit import QuantumCircuit, transpile
import qiskit.pulse as pulse
from qiskit.circuit import Parameter
from qiskit_ibm_runtime.fake_provider import FakeValenciaV2
from qiskit_experiments.library import QubitSpectroscopy, Rabi, T1, T2Ramsey
from qiskit_experiments.framework import ParallelExperiment
from qiskit_experiments.calibration_management import Calibrations
import qiskit_aer as aer
from qiskit_aer.primitives import Estimator
from qiskit_experiments.library import QubitSpectroscopy
```

```python
backend = FakeValenciaV2()
qubit = 0
estimated_freq = backend.target.qubit_properties[qubit].frequency

center_freq = estimated_freq
freq_span = 40e6 # 40 MHz
num_points = 51

frequencies = np.linspace(center_freq - freq_span/2, 
                          center_freq + freq_span/2, 
                          num_points)


sim_backend = aer.AerSimulator.from_backend(backend)
if sim_backend.options.get("hamiltonian") is None:
    print("警告：后端没有加载哈密顿量，脉冲模拟可能无法运行。")

spec_exp = QubitSpectroscopy(physical_qubits=[0], frequencies=frequencies)

exp_data = spec_exp.run(backend = sim_backend).block_for_results()

if len(exp_data.data()) > 0:
    print(f"✅ 成功获取了 {len(exp_data.data())} 条数据点！")
    display(exp_data.figure(0))
else:
    print("❌ 数据依然为空。尝试最后的‘手动降级’大法...")
```
